```python
import pandas as pd
```

```python
df = pd.read_csv('wales.csv')
df
```

```python
filtered_df = df[~df['status'].isin(['canceled', 'paused'])]
filtered_df
```

```python
result_df = filtered_df.groupby('customer_name').sum()
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('wales.xlsx'):
    df.to_excel('wales.xlsx', sheet_name='raw_data')
    filtered_df.to_excel('wales.xlsx', sheet_name='filtered_data')
    result_df.to_excel('wales.xlsx', sheet_name='result_data')
```
