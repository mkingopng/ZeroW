```python
import pandas as pd
```

```python
df = pd.read_csv('sunshine_coast.csv')
df
```

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

```python
result_df = filtered_df.groupby('CUSTOMER')[['NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING']].sum()
result_df
```

```python
result_df = result_df[result_df['NEXT_PAYMENT'] != 0]
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('sunshine_coast.xlsx') as writer:
    df.to_excel(writer, sheet_name='raw_data')
    filtered_df.to_excel(writer, sheet_name='filtered_data')
    result_df.to_excel(writer, sheet_name='membership_analysis')
```
