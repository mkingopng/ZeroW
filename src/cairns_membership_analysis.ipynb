```python
import pandas as pd
```

```python
df = pd.read_csv('cairns.csv')
df
```

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

```python
result_df = filtered_df.groupby('CUSTOMER')[['NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING']].sum()
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('cairns.xlsx') as writer:
    df.to_excel(writer, sheet_name='original data')
    filtered_df.to_excel(writer, sheet_name='total payments')
    result_df.to_excel(writer, sheet_name='unique members')
```
