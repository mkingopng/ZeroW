```python
import pandas as pd
```

```python
df = pd.read_csv('brisbane.csv')
df
```

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

```python
result_df = filtered_df.groupby('CUSTOMER')[['NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING']].sum()
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('brisbane.xlsx') as writer:
    df.to_excel(writer, sheet_name='Raw Data')
    filtered_df.to_excel(writer, sheet_name='Number of transactions')
    result_df.to_excel(writer, sheet_name='Unique Members')
```
