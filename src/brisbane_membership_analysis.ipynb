```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_csv('brisbane.csv')
df
```

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

```python
result_df = filtered_df.groupby('CUSTOMER').sum()
result_df
```

```python
result_df = result_df[result_df['NEXT_PAYMENT'] != 0]
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('brisbane.xlsx') as writer:
    df.to_excel(writer, sheet_name='Raw Data', index=False)
    filtered_df.to_excel(writer, sheet_name='Number of transactions', index=False)
    result_df.to_excel(writer, sheet_name='Unique Members', index=False)
```
