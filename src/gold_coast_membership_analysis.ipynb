```python
import pandas as pd
```

```python
df = pd.read_csv('gold_coast.csv')
df
```

```python
filtered_df = df[~df['status'].isin(['canceled', 'paused'])]
filtered_df
```

```python
result_df = filtered_df.groupby('customer_name').sum()
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('gold_coast.xlsx') as writer:
    df.to_excel(writer, sheet_name='Raw Data', index=False)
    filtered_df.to_excel(writer, sheet_name='Number of transactions', index=False)
    result_df.to_excel(writer, sheet_name='Unique Members', index=False)
```
