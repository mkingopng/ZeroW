```python
import pandas as pd
```

```python
df = pd.read_csv('southside.csv')
df
```

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused'])]
filtered_df
```

```python
filtered_df = filtered_df[~filtered_df['TYPE'].isin(['One-off', 'Fixed'])]
filtered_df
```

```python
result_df = filtered_df.groupby('CUSTOMER')[['NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING']].sum()
result_df
```
