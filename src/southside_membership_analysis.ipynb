```python
import pandas as pd
```

```python
df = pd.read_csv('southside.csv')
df
```

### filter out 'Cancelled', 'Paused', 'One-off', 'Fixed' from 'STATUS'

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

### group by 'CUSTOMER' and sum 'NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING'

```python
groupby_df = filtered_df.groupby('CUSTOMER')[['NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING']].sum()
groupby_df
```

```python
groupby_df.index.nunique()
```

```python
with pd.ExcelWriter('southside.xlsx') as writer:
    df.to_excel(writer, sheet_name='original')
    filtered_df.to_excel(writer, sheet_name='filtered')
    groupby_df.to_excel(writer, sheet_name='groupby')
```
