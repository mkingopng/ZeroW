```python
import pandas as pd
```

```python
df = pd.read_csv('southside.csv')
df
```

### filter out 'Cancelled', 'Paused', 'One-off', 'Fixed' from 'STATUS'

```python
filtered_df = df[~df['STATUS'].isin(['Cancelled', 'Paused', 'One-off', 'Fixed'])]
filtered_df
```

### group by 'CUSTOMER' and sum 'NEXT_PAYMENT', 'LAST_PAYMENT', 'TOTAL_VALUE', 'PAID', 'REMAINING'

```python
result_df = filtered_df.groupby('CUSTOMER').sum()
result_df
```

```python
result_df = result_df[result_df['NEXT_PAYMENT'] != 0]
result_df
```

```python
result_df.index.nunique()
```

```python
with pd.ExcelWriter('southside.xlsx') as writer:
    df.to_excel(writer, sheet_name='original', index=False)
    filtered_df.to_excel(writer, sheet_name='filtered', index=False)
    result_df.to_excel(writer, sheet_name='groupby', index=False)
```
